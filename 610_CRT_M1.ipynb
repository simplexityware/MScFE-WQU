1. Write the CAPM the equation, specifying what each term means.

> The CAPM describes the relationship between systematic risk and expected return for assets(stocks in this exercise)

>> $E(R_i)=R_f+\beta_i(E(R_m)-R_f)$

> where:

>> $E(R_i)$ is the expected return on an $i$nvestment

>> $E(R_m)$ is the expected $m$arket return

>> $\beta_i$ is the systemic or market risk of the $i$nvestment

>> $R_f$ is the risk-free rate of return


2. Explain in words how the model explains the return of a stock.  

> The above equation can be simplified as follows:

>> $R_i=R_f+\beta_i(R_m-R_f)$

> where:

>> $R_i$ is the rate of return of the $i$nvestment

>> $R_m$ is the rate of return of the $m$arket

> Rearranging the above equation to provide an interpretation as a regression line called the $S$ecurity $M$arket $L$ine: 

>> $R_i - R_f = \alpha + \beta_i(R_m-R_f) + \epsilon$

> where:

>> $R_i - R_f$ is the $i$nvestment risk premium

>> $R_m - R_f$ is the $m$arket risk premium

> Therefore, the equation can be interpreted as: the risk premium of individual stock depends on its market risk, $\beta_i$ and the market risk premium, $(R_m-R_f)$. $\beta_i$, representing the slope of the preceding linear regression, measures the increase of the expected market risk premium as the individual stock's expected return increases and vice versa, ie when the former increases by 1%, the latter increases by $\beta_i$%

8. Formulate the CAPM using your stockâ€™s return, the benchmark return, and a risk-free rate. 


```python
df_13wkTBill = getData.get_data_yahoo('^IRX', dt_begdate, dt_enddate)
df_13wkTBill
```

```python
df_13wkTBill['Risk-Free'] = (1 + df_13wkTBill['Adj Close'] / 100) ** (13 * 5 / 252) - 1
df_13wkTBill
```

```python
#from pandas.stats.api import ols
import statsmodels.api as sm
df_y = df_aapl['Returns'] - df_13wkTBill['Risk-Free']
df_x = df_sp500['Returns'] - df_13wkTBill['Risk-Free']
#result = ols(y=df_y, x=df_x)
results = sm.OLS(df_y[200:], df_x[200:]).fit()
print(results.summary())
```
